```python
def feature_engineering(df) :
    # Nan
    df = df.fillna(0.0)
    
    # datetime object
    df['datetime'] = df['date'] + ", " + df['time']
    df = df.drop(['date', 'time'], axis = 1)
    df['datetime'] = pd.to_datetime(df['datetime'], format="%Y-%m-%d, %H:%M:%S")
    
    return df
```

```python
def splitById(id, df) :
    idx = df['store_id'] == id
    split = df[idx]
    return split

split_0 = splitById(0, Raw_train)
split_0
```
