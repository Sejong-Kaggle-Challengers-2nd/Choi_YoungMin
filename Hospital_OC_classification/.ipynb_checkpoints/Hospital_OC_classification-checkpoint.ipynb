# 데이터 로더

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 100)
```

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
submission_sample = pd.read_csv("submission_sample.csv")
```

# 데이터 전처리

## train data

```python
train_data.head()
```

분류 성능에 영향을 미치지 않을 것으로 예상되는 feature들
1. inst_id ; 각 파일에서의 병원 고유 번호
    - 파일에서의 병원 고유 번호는 병원 개폐 여부와는 관련 없는 Data라고 판단됨.
2. openDate ; 병원의 설립일
    - 병원이 언제 폐업했는지에 대한 feture가 없으므로 관련이 없다고 판단됨.

```python
# 분류 성능과 관련없다고 판단된 feature를 drop
train_data = train_data.drop(['inst_id', 'openDate'], axis=1)
```

```python
# 결측 data 제거
train_data = train_data.dropna()
```

```python
# input data와 label data 분할
X_train = train_data.drop(['OC'], axis=1)
print(X_train.shape)

y = train_data['OC'].values
print(y.size)
```

실수화 해야하는 범주형 data들
1. label date ; y_train
2. sido feature
3. instkind feature
4. ownerChange feature

```python
# 범주형 data 실수화

from sklearn.preprocessing import LabelEncoder

classle = LabelEncoder()

y_train = classle.fit_transform(y)

print("Encoded label data : ", np.unique(y_train))
```

```python
# sido feature 실수화
tmp = classle.fit_transform(X_train['sido'])
X_train['sido'] = tmp

# instkind feature 실수화
tmp = classle.fit_transform(X_train['instkind'])
X_train['instkind'] = tmp

# ownerChange feature 실수화
tmp = classle.fit_transform(X_train['ownerChange'])
X_train['ownerChange'] = tmp

X_train
```

```python
X_train = X_train.astype(np.float64)
```

```python
X_train
```

## test data

```python
test_data.head()
```

```python
# 분류 성능과 관련없다고 판단된 feature와 label data를 drop
X_test = test_data.drop(['inst_id', 'OC', 'openDate'], axis=1)
```

```python
# 결측 data 제거
X_test = X_test.dropna()
```

```python
# 범주형 data 실수화

# sido feature 실수화
tmp = classle.fit_transform(X_test['sido'])
X_test['sido'] = tmp

# instkind feature 실수화
tmp = classle.fit_transform(X_test['instkind'])
X_test['instkind'] = tmp

# ownerChange feature 실수화
tmp = classle.fit_transform(X_test['ownerChange'])
X_test['ownerChange'] = tmp

X_test
```

```python
X_test['employee2'] = tmp

X_test['employee2'] = pd.to_numeric(X_test['employee2'])
```

```python
X_test = X_test.astype(np.float64)
```

## 입력 데이터 표준화

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

# Traning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5, p = 2)
knn.fit(X_train_std, y_train)

y_train_pred_knn = knn.predict(X_train_std)
predict_knn = knn.predict(X_test_std)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train, y_train_pred_knn))
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(penalty = 'l2', C = 200, random_state=1, max_iter=500)
lr.fit(X_train_std, y_train)

y_train_pred_lr = lr.predict(X_train_std)
predict_lr = lr.predict(X_test_std)
```

```python
print(accuracy_score(y_train, y_train_pred_lr))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(criterion = 'entropy', max_depth = 9, random_state = 1)
dtc.fit(X_train_std, y_train)

y_train_pred_dtc = dtc.predict(X_train_std)
predict_dtc = dtc.predict(X_test_std)
```

```python
print(accuracy_score(y_train, y_train_pred_dtc))
```

## LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
lda.fit(X_train_std, y_train)

y_train_pred_lda = lda.predict(X_train_std)
predict_lda = lda.predict(X_test_std)
```

```python
print(accuracy_score(y_train, y_train_pred_lda))
```

## QDA

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

qda = QuadraticDiscriminantAnalysis()
qda.fit(X_train_std, y_train)

y_train_pred_qda = qda.predict(X_train_std)
predict_qda = qda.predict(X_test_std)
```

```python
print(accuracy_score(y_train, y_train_pred_qda))
```

# Submission csv file

```python
predict_knn = predict_knn.astype(np.int64)
predict_dtc = predict_dtc.astype(np.int64)
predict_lda = predict_lda.astype(np.int64)
predict_qda = predict_qda.astype(np.int64)
```

```python
submit_knn = submission_sample.copy()
submit_dtc = submission_sample.copy()
submit_lda = submission_sample.copy()
submit_qda = submission_sample.copy()
```

```python
for i in range(len(predict_knn)) :
    submit_knn['OC'][i] = predict_knn[i]
    submit_dtc['OC'][i] = predict_dtc[i]
    submit_lda['OC'][i] = predict_lda[i]
    submit_qda['OC'][i] = predict_qda[i]
```

```python
submit_knn.to_csv('submit_knn.csv', mode='w', header=True, index=False)
submit_dtc.to_csv('submit_dtc.csv', mode='w', header=True, index=False)
submit_lda.to_csv('submit_lda.csv', mode='w', header=True, index=False)
submit_qda.to_csv('submit_knn.qda', mode='w', header=True, index=False)
```
